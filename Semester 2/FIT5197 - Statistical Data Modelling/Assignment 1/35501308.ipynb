## Question 1 (7.5 marks)

Let $X\,{\sim}\,\mathcal{IG}\left(\theta: \left(\mu, \lambda\right)\right),\;\,\forall\, \mu>0 \text{ and } \lambda>0$. This means
the random varible $X$ follows the **inverse Gaussian distribution** with the set $\left(\theta: \left(\mu, \lambda\right)\right)$ acting as the parameters of said distribution. Given that we observe a sample of size $n$ that is independently and identically distributed from this distribution ([i.i.d](https://en.wikipedia.org/wiki/Independent_and_identically_distributed_random_variables)), $\mathbf{x}=$ $\left(x_{1}, \ldots, x_{n}\right)$, please find the [maximum likelihood estimate](https://en.wikipedia.org/wiki/Maximum_likelihood_estimation) for $\mu$ and $\lambda$, that is $\mu_{\text{MLE}}$ and $\lambda_{\text{MLE}}$. The probability density function (**PDF**) is as follows:

$$
f(x \mid \mu, \lambda)=\left\{\begin{array}{cc}
\left(\frac{\lambda}{2 \pi x^{3}}\right)^{1 / 2} e^{\frac{-\lambda(x-\mu)^{2}}{2 \mu^{2} x}}, & x>0 \\
0, & x \leq 0
\end{array}\right.
$$

### ANSWER

ANSWER

Likelihood Function
For independent data $x_1, x_2, \ldots, x_n$ from the inverse Gaussian distribution $\operatorname{IG}(\mu, \lambda)$, the joint likelihood is:

$$
L(\mu, \lambda)=\prod_{i=1}^n f\left(x_i \mid \mu, \lambda\right)=\prod_{i=1}^n\left(\frac{\lambda}{2 \pi x_i^3}\right)^{\frac{1}{2}} \exp \left(-\frac{\lambda\left(x_i-\mu\right)^2}{2 \mu^2 x_i}\right)
$$


Negative Log-Likelihood with "log"

$$
\begin{aligned}
\ell(\mu, \lambda) & =-\log L(\mu, \lambda) \\
& =-\sum_{i=1}^n\left[\frac{1}{2} \log \left(\frac{\lambda}{2 \pi x_i^3}\right)-\frac{\lambda\left(x_i-\mu\right)^2}{2 \mu^2 x_i}\right] \\
& =-\frac{n}{2} \log \lambda+\frac{n}{2} \log (2 \pi)+\frac{3}{2} \sum_{i=1}^n \log x_i+\frac{\lambda}{2} \sum_{i=1}^n \frac{\left(x_i-\mu\right)^2}{\mu^2 x_i}
\end{aligned}
$$


Differentiate the Negative Log-Likelihood w.r.t. $\mu$ and Find $\mu_{M L E}$
Deriving $\ell(\mu, \lambda)$ with respect to $\mu$ :

$$
\frac{\partial \ell}{\partial \mu}=\frac{\lambda}{2} \sum_{i=1}^{\mathrm{n}} \frac{\partial}{\partial \mu}\left(\frac{\left(x_i-\mu\right)^2}{\mu^2 x_i}\right)
$$


Calculating the derivative inside the sum step-by-step:

$$
\frac{\partial}{\partial \mu}\left(\frac{\left(x_i-\mu\right)^2}{\mu^2 x_i}\right)
$$

$$
\frac{d}{d \mu}\left(\frac{u}{v}\right)=\frac{u^{\prime} v-u v^{\prime}}{v^2}
$$

- $u^{\prime}=\frac{d}{d \mu}\left(x_i-\mu\right)^2=2\left(x_i-\mu\right)(-1)=-2\left(x_i-\mu\right)$
- $v^{\prime}=\frac{d}{d \mu}\left(\mu^2 x_i\right)=2 \mu x_i$

Plugging these we get

$$
\frac{\partial}{\partial \mu}\left(\frac{\left(x_i-\mu\right)^2}{\mu^2 x_i}\right)=\frac{-2\left(x_i-\mu\right) \cdot\left(\mu^2 x_i\right)-\left(x_i-\mu\right)^2 \cdot 2 \mu x_i}{\left(\mu^2 x_i\right)^2}
$$


Simplifying the numerator we get

$$
-2\left(x_i-\mu\right) \mu^2 x_i-2\left(x_i-\mu\right)^2 \mu x_i=-2 x_i \mu^2\left(x_i-\mu\right)-2 x_i \mu\left(x_i-\mu\right)^2
$$


So the full simplified answer is:

$$
\frac{\partial}{\partial \mu}\left(\frac{\left(x_i-\mu\right)^2}{\mu^2 x_i}\right)=\frac{-2 x_i \mu^2\left(x_i-\mu\right)-2 x_i \mu\left(x_i-\mu\right)^2}{\left(\mu^2 x_i\right)^2}
$$

WKT

$$
\frac{\partial \ell}{\partial \mu}=\frac{\lambda}{2} \sum_{i=1}^{\mathrm{n}} \frac{\partial}{\partial \mu}\left(\frac{\left(x_i-\mu\right)^2}{\mu^2 x_i}\right)
$$

From 
$$
\frac{\partial}{\partial \mu}\left(\frac{\left(x_i-\mu\right)^2}{\mu^2 x_i}\right)=\frac{-2 x_i \mu^2\left(x_i-\mu\right)-2 x_i \mu\left(x_i-\mu\right)^2}{\left(\mu^2 x_i\right)^2}
$$

Factor out $-2 x_i$ from the numerator:

$$
-2 x_i\left[\mu^2\left(x_i-\mu\right)+\mu\left(x_i-\mu\right)^2\right]
$$

Expand $\mu\left(x_i-\mu\right)^2=\mu\left(x_i^2-2 x_i \mu+\mu^2\right)=\mu x_i^2-2 \mu^2 x_i+\mu^3$

Now add $\mu^2\left(x_i-\mu\right)$ :

$$
\mu^2 x_i-\mu^3+\mu x_i^2-2 \mu^2 x_i+\mu^3=\left(\mu^2 x_i-2 \mu^2 x_i\right)+\left(\mu x_i^2\right)+\left(-\mu^3+\mu^3\right)=-\mu^2 x_i+\mu x_i^2
$$


So, the numerator becomes:

$$
-2 x_i\left[-\mu^2 x_i+\mu x_i^2\right]=2 x_i\left[\mu^2 x_i-\mu x_i^2\right]=2 x_i^2 \mu^2-2 x_i^3 \mu
$$


Thus,

$$
\frac{\partial}{\partial \mu}\left(\frac{\left(x_i-\mu\right)^2}{\mu^2 x_i}\right)=\frac{2 x_i^2 \mu^2-2 x_i^3 \mu}{\left(\mu^2 x_i\right)^2}
$$


Now, sum over $i$, and include the prefactor $\frac{\lambda}{2}$ :
Final Step

$$

Simplifying denominator and numerator by dividing numerator and denominator by $x_i^2$ (since $x_i \neq 0$ ):

=\lambda \sum_{i=1}^n \frac{\mu^2-x_i \mu}{\mu^4}
$$

$$

Set $\frac{\partial \ell}{\partial \mu}=0$ and by algebraic manipulation, the maximizing estimate is:

$$
\frac{\partial \ell}{\partial \mu}=\lambda \sum_{i=1}^n \frac{\mu-x_i}{\mu^3}=0

Since $\lambda$ and $\mu^3$ (assuming $\mu \neq 0$ ) are not zero, divide both sides by $\lambda$ :

Expanding the sum:

$$
\begin{gathered}

n \mu-\sum_{i=1}^n x_i=0 \\
n \mu=\sum_{i=1}^n x_i \\
\mu=\frac{1}{n} \sum_{i=1}^n x_i
\end{gathered}
$$


So, the maximum likelihood estimate (MLE) for $\mu$ is simply the sample mean:

$$
\hat\mu_{\text {MLE }}=\frac{1}{n} \sum_{i=1}^n x_i

Differentlating with respect to $\lambda$ and setting to zero, we get

$$
\frac{\partial l}{\partial \lambda}=-\frac{n}{2 \lambda}+\frac{1}{2} \sum_{i=1}^n \frac{\left(x_i-\mu\right)^2}{\mu^2 x_i}=0

Multiply both sides by 2 :

\frac{n}{\lambda}=\sum_{i=1}^n \frac{\left(x_i-\mu\right)^2}{\mu^2 x_i} \\

Split the terms

\frac{x_i^2}{\bar{x}^2 x_i}=\frac{x_i}{\bar{x}^2} \quad \frac{-2 x_i \bar{x}}{\bar{x}^2 x_i}=\frac{-2}{\bar{x}} \quad \frac{\bar{x}^2}{\bar{x}^2 x_i}=\frac{1}{x_i}

## Question 2 (7.5 marks)
Suppose that we know that the random variable $X{\,\sim\,}\mathcal{Dist}(\mu=\theta, \sigma^2=\theta^2)$ follows the PDF given below:

$$
\begin{equation}
f\left(x|\theta\right)=
\begin{cases}
\frac{1}{\theta} \exp(-\frac{x}{\theta}) & x > 0 \\
0 & \text{otherwise.}
\end{cases}
\end{equation}
$$

Given a sample of $n$ [i.i.d](https://en.wikipedia.org/wiki/Independent_and_identically_distributed_random_variables) observations $\mathbf{x}=$ $\left(x_{1}, \ldots, x_{n}\right)$ from this distribution, please answer the following questions:

*(a)* Derive the MLE estimator for $\theta$, i.e., $\hat{\theta}_{\text{MLE}}$, and show that it is unbiased. [2.5 Marks]

*(b)* Find an estimator with better MSE (i.e smaller MSE) compared to the $\hat{\theta}_{\text{MLE}}$ obtained from (a). [5 Marks]

### ANSWER

### (a) 
MLE Estimator Derivation

Likelihood Function
Given the PDF:

$$
f(x \mid \theta)= \begin{cases}\frac{1}{\theta} \exp \left(-\frac{x}{\theta}\right), & x>0 \\ 0, & \text { otherwise }\end{cases}
$$


The likelihood for i.i.d. sample $\mathbf{x}=\left(x_1, \ldots, x_n\right)$ is:

$$
L(\theta)=\prod_{i=1}^n \frac{1}{\theta} \exp \left(-\frac{x_i}{\theta}\right)=\frac{1}{\theta^n} \exp \left(-\frac{1}{\theta} \sum_{i=1}^n x_i\right)
$$


Negative Log-Likelihood
Taking the negative log of the likelihood:

$$
-\ell(\theta)=n \log \theta+\frac{1}{\theta} \sum_{i=1}^n x_i
$$


Differentiate w.r.t. $\theta$

$$
\frac{d}{d \theta}(-\ell(\theta))=\frac{n}{\theta}-\frac{1}{\theta^2} \sum_{i=1}^n x_i
$$


Set derivative to zero

$$
\frac{n}{\hat{\theta}}-\frac{1}{\hat{\theta}^2} \sum_{i=1}^n x_i=0
$$


Multiplying both sides by $\hat{\theta}^2$ :

$$
n \hat{\theta}-\sum_{i=1}^n x_i=0
$$


Solve for $\hat{\theta}$

$$
\hat{\theta}=\frac{1}{n} \sum_{i=1}^n x_i=\bar{x}
$$


Proof of Unbiasedness
Since $\mathbb{E}\left[X_i\right]=\theta$, it follows that:

$$
\mathbb{E}[\hat{\theta}]=\mathbb{E}[\bar{x}]=\theta
$$


Hence, $\hat{\theta}$ is unbiased.

### (b)
Construction of an Improved Estimator with Smaller MSE

Variance of $\hat{\theta}$
Given the variance of each $X_i$ is $\theta^2$, the variance of the sample mean is:

$$
\operatorname{Var}(\hat{\theta})=\operatorname{Var}\left(\frac{1}{n} \sum_{i=1}^n x_i\right)
$$

If you multiply a random variable by a constant, the variance is multiplied by the square of that constant. Here, the constant is $\frac{1}{n}$, so the variance is divided by $n^2$.

$$
=\frac{1}{n^2} \operatorname{Var}\left(\sum_{i=1}^n x_i\right)
$$
$$
=\frac{1}{n^2} \sum_{i=1}^n \operatorname{Var}\left(x_i\right)
$$


Every $ x_i $ has variance $\theta^2$, so you sum this up $n$ times, giving $ n\theta^2 $.

$$
=\frac{1}{n^2} \cdot n \theta^2
$$
$$
=\frac{\theta^2}{n} .
$$

$$
\operatorname{Var}(\hat{\theta})=\operatorname{Var}(\bar{x})=\frac{\theta^2}{n}
$$


Define a scaled estimator
Consider:

$$
\tilde{\theta}=c \bar{x}
$$

where $c$ is a scalar parameter.

$$
\mathbb{E}[\hat{\theta}]=c \mathbb{E}[\bar{x}]=c .\theta,
$$
$$
\operatorname{Bias}(\bar{x})=\mathbb{E}[\bar{x}]-\theta
$$
$$
\quad \operatorname{Bias}(\bar{\theta})=c \theta-\theta
$$

Calculate MSE of $\tilde{\theta}$
MSE is:

$$
\begin{aligned}
\operatorname{MSE}(\tilde{\theta}) & =\operatorname{Var}(\bar{\theta})+\operatorname{Bias}(\bar{\theta})^2 \\
& =c^2 \frac{\theta^2}{n}+(c \theta-\theta)^2
\end{aligned}
$$
$$
=c^2 \frac{\theta^2}{n}+(c \theta-\theta)^2
$$


Expanding $(c \theta-\theta)^2$

$$
(c \theta-\theta)^2=[c \theta-1 \cdot \theta]^2=[\theta(c-1)]^2=\theta^2(c-1)^2
$$


Substituting it back, we get

$$
=c^2 \frac{\theta^2}{n}+\theta^2(c-1)^2
$$


Taking Out $\theta^2$, we get

$$
=\theta^2\left(\frac{c^2}{n}+(c-1)^2\right)
$$


Therefore

$$
=\theta^2\left(\frac{c^2}{n}+(c-1)^2\right)
$$


Minimize MSE w.r.t. $c$
Take derivative:

$$
\frac{d}{d c}\left(\frac{c^2}{n}+(c-1)^2\right)=\frac{2 c}{n}+2(c-1)=0
$$


Solving for $c$ :

$$
c\left(1+\frac{1}{n}\right)=1 \quad \Longrightarrow \quad c=\frac{n}{n+1}
$$


Improved estimator and comparison of MSE
The improved estimator is:

$$
\dot{\theta}=\frac{n}{n+1} \bar{x}
$$


Computing its MSE:

$$
\operatorname{MSE}(\bar{\theta})=\theta^2\left(\frac{c^2}{n}+(c-1)^2\right)=\theta^2 \cdot \frac{1}{n+1}<\frac{\theta^2}{n}=\operatorname{MSE}(\hat{\theta})
$$


Therefore, $\tilde{\theta}$ has smaller mean squared error than the MLE $\hat{\theta}$.

## Question 4 (7.5 marks)
Suppose that we know that the random variable $X$ follows the PDF given below:

$$
\begin{equation}
    f\left(x|\theta\right)= 
    \begin{cases}
         e^{-(x-\theta)} & x\geq\theta \\
         0 & \text{otherwise.}
    \end{cases}
\end{equation}
$$

Given a sample of $n$ [i.i.d](https://en.wikipedia.org/wiki/Independent_and_identically_distributed_random_variables) observations $\mathbf{x}=$ $\left(x_{1}, \ldots, x_{n}\right)$ from this distribution, please answer the following questions:

**(a)** Derive the MLE estimator for $\theta$, i.e., $\hat{\theta}_{\text{MLE}}$. [4.5 Marks]

**(b)** Show that the estimator $\hat{\theta} = \overline{X} - 1$ (where $\overline{X} = \frac{1}{n}\sum_{i=1}^n X_i$) is an unbiased and consistent estimator for the given distribution. [3 Marks]

### ANSWER

### (a)
Given the PDF is,

$$
f(x ; \theta)= \begin{cases}e^{-(x-\theta)} & x \geq \theta \\ 0 & \text { otherwise }\end{cases}
$$


For i.i.d. observations $x_1, x_2, \ldots, x_n$ :

$$
L(\theta)=\prod_{i=1}^n f\left(x_i ; \theta\right)=\prod_{i=1}^n e^{-\left(x_i-\theta\right)} \cdot I\left(x_i \geq \theta\right)
$$

where $I\left(x_i \geq \theta\right)$ is the indicator function, 1 if true, 0 otherwise.

The (natural) log-likelihood is:

$$
\ell(\theta)=\log L(\theta)=\sum_{i=1}^n \log \left[e^{-\left(x_i-\theta\right)}\right]=\sum_{i=1}^n\left[-\left(x_i-\theta\right)\right]
$$


Thus,

$$
\ell(\theta)=-\sum_{i=1}^n x_i+n \theta
$$

for $\theta \leq \min x_i$, and $-\infty$ otherwise.

The negative log-likelihood is:

$$
-\ell(\theta)=\sum_{i=1}^n x_i-n \theta
$$

for $\theta \leq \min x_i$.

Because $\sum_{i=1}^n x_i$ is constant, $-n \theta$ is minimized for the largest feasible $\theta$.
But, by the support constraint, $\theta \leq \min \left(x_1, \ldots, x_n\right)$.
Thus, the negative log-likelihood is minimized when:

$$
\theta=\min \left(x_1, x_2, \ldots, x_n\right)
$$



So, the MLE for $\boldsymbol{\theta}$ using the negative log-likelihood is:

$$
\hat{\theta}_{M L E}=\min \left(x_1, x_2, \ldots, x_n\right)
$$

### (b)
Given:

$$
f(x \mid \theta)= \begin{cases}e^{-(x-\theta)} & x \geq \theta \\ 0 & \text { otherwise }\end{cases}
$$


Let

$$
Y=X-\theta
$$


This implies that

$$
Y \sim \text { Exponential }(1), \quad f_Y(y)=e^{-y}, \quad y \geq 0
$$


Writing $\bar{X}$ in terms of $\bar{Y}$
The sample mean is:

$$
\bar{X}=\frac{1}{n} \sum_{i=1}^n X_i=\frac{1}{n} \sum_{i=1}^n\left(Y_i+\theta\right)=\bar{Y}+\theta
$$

where

$$
\bar{Y}=\frac{1}{n} \sum_{i=1}^n Y_i
$$


Calculate the expectation $E[\hat{\theta}]$
The estimator is defined as:

$$
\hat{\theta}=\bar{X}-1=(\bar{Y}+\theta)-1=\bar{Y}+(\theta-1)
$$


Use the linearity of expectation:

$$
E[\hat{\theta}]=E[\bar{Y}]+\theta-1
$$


Since all $Y_i$ are i.i.d. with exponential(1):

$$
E[\bar{Y}]=E\left[Y_i\right]=1
$$


Therefore,

$$
E[\hat{\theta}]=1+\theta-1=\theta
$$


So, $\hat{\theta}$ is unbiased because its expectation equals $\hat{\theta}$.
Step 4: Show consistency $\hat{\theta} \xrightarrow{p} \theta$
By the Law of Large Numbers:

$$
\bar{Y} \xrightarrow{P} E[Y]=1 \quad \text { as } n \rightarrow \infty
$$


Therefore,

$$
\hat{\theta}=\bar{Y}+(\theta-1) \xrightarrow{p} 1+(\theta-1)=\theta

So, $\hat{\theta}$ is consistent.

-\frac{2}{7} x^2+x+\frac{1}{8}-p
$$


Rearranged

$$
\frac{2}{7} x^2-x+\left(p-\frac{1}{8}\right)=0
$$


Quadratic formula is

$$
x=\frac{1 \pm \sqrt{1-\frac{4}{4}\left(p-\frac{1}{4}\right)}}{\frac{4}{7}}-\frac{7}{4}\left(1 \pm \sqrt{1-\frac{8}{7}\left(p-\frac{1}{8}\right)}\right)
$$


Simplifying Inside the root,

$$
1-\frac{8}{7}\left(p-\frac{1}{8}\right)-\frac{8}{7}(1-p)
$$


Taking root in $\left[0, \frac{7}{4}\right]$ 

$$
Q(p)=\frac{7}{4}\left(1-\sqrt{\frac{8}{7}(1-p)}\right)
$$


Final Expression for Quantile Function

$$
Q(p)= \begin{cases}\frac{-1+\sqrt{8p}}{4}, & 0 \leq p \leq \frac{1}{8} \\ \frac{7}{4}\left(1-\sqrt{\frac{8}{7}(1-p)}\right), & \frac{1}{8}<p \leq 1\end{cases}
$$

````r
samplingHelper <- function(n) {
  #--------------------------
  # Error handling: check input
  #--------------------------
  if (!is.numeric(n) || length(n) != 1 || n <= 0 || n != round(n)) {
    stop("n must be a positive integer.")
  }
  
  #--------------------------
  # Generate n Uniform[0,1] samples
  #--------------------------
  p <- runif(n)
  
  #--------------------------
  # Preallocate output vector
  #--------------------------
  numbers <- numeric(n)
  
  #--------------------------
  # Compute quantiles for p <= 1/8 and p > 1/8
  #--------------------------
  # Logical indices
  idx1 <- p <= 1/8
  idx2 <- !idx1
  
  # Apply quantile formulas
  numbers[idx1] <- (-1 + sqrt(8 * p[idx1])) / 4
  numbers[idx2] <- (7/4) * (1 - sqrt((8/7) * (1 - p[idx2])))
  
  return(numbers) # numbers is now a vector of samples from f(x)
}
````

### (d)
### Detailed Summary and Justification of the Sampling Method

The sampling method employed is based on the inverse transform technique, which utilizes the quantile function to generate samples from the target distribution.

1. **Derivation of Quantile Function:**  
   Starting from the given probability density function \( f(x) \), the cumulative distribution function (CDF) \( F(x) \) is derived by integrating \( f(x) \) piecewise. Subsequently, the inverse CDF or quantile function \( Q(p) \) is obtained by solving \( F(x) = p \) for \( x \). The quantile function has the form:
   $$
   Q(p) =
   \begin{cases}
   \frac{-1 + \sqrt{8p}}{4}, & 0 \leq p \leq \frac{1}{8} \\[2em]
   \frac{7}{4}\left(1 - \sqrt{\frac{8}{7}(1-p)}\right), & \frac{1}{8} < p \leq 1
   \end{cases}
   $$

2. **Generation of Uniform Random Variables:**  
   \( n \) independent uniform random variables \( p \) are generated from the uniform distribution on [0, 1].

3. **Transformation Using Quantile Function:**  
   Each uniform variable \( p \) is transformed to a value \( x = Q(p) \), where the appropriate form of \( Q(p) \) is selected based on the value of \( p \).

**Justification:**  
The quantile function \( Q(p) \) gives the value below which a proportion \( p \) of data falls under the distribution \( f(x) \). Generating uniform random numbers \( p \) over [0, 1] ensures that all cumulative probabilities are equally likely. By transforming these according to \( Q(p) \), samples are produced that correctly represent the distribution of \( f(x) \). This method leverages the one-to-one relationship between the uniform distribution and any continuous distribution via their CDFs and provides an exact and efficient sampling technique. It is a standard approach widely adopted in simulation and Monte Carlo methods when the quantile function can be explicitly computed.


### (e)
Verify it is a valid PDF
- Check normalization:

$$
\int_{-\infty}^{\infty} e^{-\pi x^2} d x=1
$$


Recall that the Gaussian integral:

$$
\int_{-\infty}^{\infty} e^{-a x^2} d x=\sqrt{\frac{\pi}{a}}
$$


With $a=\pi$, this integral equals $\sqrt{\frac{\pi}{\pi}}=1$, so $f(x)$ is a valid PDF.

Deriving the CDF $F(x)$
The CDF is defined as:

$$
F(x)=P(X \leq x)=\int_{-\infty}^x e^{-\pi t^2} d t
$$


Expressing $F(x)$ using the error function
The error function is defined as:

$$
\operatorname{erf}(z)=\frac{2}{\sqrt{\pi}} \int_0^z e^{-t^2} d t
$$


To rewrite our CDF in terms of erf, perform the substitution:
- Let $u=t \sqrt{\pi} \Rightarrow t=\frac{u}{\sqrt{\pi}}$
- Therefore, $d t=\frac{1}{\sqrt{\pi}} d u$

Change the limits accordingly:
- When $t=-\infty, u=-\infty$
- When $t=x, u=x \sqrt{\pi}$

Rewrite the integral:

$$
F(x)=\int_{-\infty}^x e^{-\pi t^2} d t=\int_{-\infty}^{x \sqrt{\pi}} e^{-u^2} \frac{1}{\sqrt{\pi}} d u=\frac{1}{\sqrt{\pi}} \int_{-\infty}^{x \sqrt{\pi}} e^{-u^2} d u
$$


Split the integral at zero:

$$
\int_{-\infty}^{x \sqrt{\pi}} e^{-u^2} d u=\int_{-\infty}^0 e^{-u^2} d u+\int_0^{x \sqrt{\pi}} e^{-u^2} d u
$$


We know the Gaussian integral

$$
\int_{-\infty}^0 e^{-u^2} d u=\frac{\sqrt{\pi}}{2}
$$


Hence,

$$
F(x)=\frac{1}{\sqrt{\pi}}\left(\frac{\sqrt{\pi}}{2}+\int_0^{x \sqrt{\pi}} e^{-u^2} d u\right)=\frac{1}{2}+\frac{1}{\sqrt{\pi}} \int_0^{x \sqrt{\pi}} e^{-u^2} d u
$$

Express the remaining integral with erf:

$$
\operatorname{erf}(z)=\frac{2}{\sqrt{\pi}} \int_0^z e^{-t^2} d t \Longrightarrow \int_0^z e^{-t^2} d t=\frac{\sqrt{\pi}}{2} \operatorname{erf}(z) .
$$


Therefore,

$$
F(x)=\frac{1}{2}+\frac{1}{\sqrt{\pi}} \cdot \frac{\sqrt{\pi}}{2} \operatorname{erf}(x \sqrt{\pi})=\frac{1}{2}+\frac{1}{2} \operatorname{erf}(x \sqrt{\pi}) .
$$


Final CDF:

$$
F(x)=\frac{1}{2}(1+\operatorname{erf}(x \sqrt{\pi})) .
$$


Deriving the Quantile Function $Q(p)=F^{-1}(p)$
Given that

$$
p=F(x)=\frac{1}{2}(1+\operatorname{erf}(x \sqrt{\pi})),
$$

solving for $\boldsymbol{x}$,

Multiplying both sides by 2, we get

$$
2 p=1+\operatorname{erf}(x \sqrt{\pi})
$$

Subtracting 1, we get

$$
2 p-1=\operatorname{erf}(x \sqrt{\pi})
$$

Applying the inverse error function,

$$
x \sqrt{\pi}=\operatorname{erf}^{-1}(2 p-1)
$$

Isolating $\boldsymbol{x}$, we get

$$
Q(p)=x=\frac{1}{\sqrt{\pi}} \operatorname{erf}^{-1}(2 p-1)
$$

````r
samplingHelper <- function(n) {
  # Validate input
  if (missing(n) || !is.numeric(n) || n <= 0 || n != floor(n)) {
    stop("Please provide a positive integer for 'n'")
  }
  
  # Generate uniform samples
  u <- runif(n)
  
  # Initialize output vector
  x <- numeric(n)
  
  # For samples where 2p-1 < 0 => erf^{-1}(2p-1) < 0, corresponding to negative x
  # For p in [0, 0.5), erf^{-1}(2p-1) < 0
  idx_neg <- u < 0.5
  # For p in [0.5, 1], erf^{-1}(2p-1) >= 0
  idx_pos <- !idx_neg
  
  # Compute for negative region
  x[idx_neg] <- - (1 / sqrt(pi)) * (qnorm(2 * u[idx_neg]) / sqrt(2))
  # Compute for positive region
  x[idx_pos] <- (1 / sqrt(pi)) * (qnorm(2 * u[idx_pos] - 1) / sqrt(2))
  
  return(x)
}
````
