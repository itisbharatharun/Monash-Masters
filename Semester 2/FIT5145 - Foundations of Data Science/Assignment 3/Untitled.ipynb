```r
library(dplyr)
library(ggplot2)

# ... Assume verstappen_year_results creation as before ...

# Calculate mean and sd per year
yearly_consistency <- verstappen_year_results %>%
  group_by(year) %>%
  summarise(mean_position = mean(position, na.rm=TRUE),
            sd_position = sd(position, na.rm=TRUE),
            races = n())

print(yearly_consistency) # Table output

# Improved Plot: show all years from 2015 to 2024
ggplot(yearly_consistency, aes(x=year, y=mean_position)) +
  geom_line(color='blue', size=1.5) +
  geom_point(color='blue', size=3) +
  geom_errorbar(aes(ymin=mean_position - sd_position, ymax=mean_position + sd_position), width=0.2, color='gray') +
  geom_text(aes(label=round(mean_position,1)), vjust=-1, size=5, color='darkblue') +
  scale_x_continuous(breaks=2015:2024, limits=c(2015,2024)) +
  scale_y_reverse(breaks=seq(1, max(yearly_consistency$mean_position, na.rm=TRUE), by=1)) + # Lower positions are better
  labs(title='Max Verstappen Annual Performance Consistency',
       subtitle='Average race finish position per year (lower is better)',
       x='Year',
       y='Average Finish Position') +
  theme_minimal(base_size=15) +
  theme(panel.grid.major = element_line(color='gray80'),
        plot.title = element_text(face='bold', hjust=0.5),
        plot.subtitle = element_text(hjust=0.5))
```
